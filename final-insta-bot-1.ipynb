```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
driver=webdriver.Chrome(ChromeDriverManager().install())
a=driver.get('https://www.instagram.com/')
```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
from selenium.common.exceptions import NoSuchElementException
wait = WebDriverWait(driver,15) ## Wait for 10 seconds....
```

# 1.1 Login to instagram 

```python
wait = WebDriverWait(driver,15) ## Wait for 10 seconds....


def Login(usename,pasword):
    
    username = driver.find_element_by_name("username")
    password = driver.find_element_by_name("password")
    
    username.send_keys(usename)
    password.send_keys(pasword)
    enter = driver.find_element_by_xpath("//button[contains(@class,'sqdOP')]/div")
    enter.click()
    
    #Not saving the info
    
    not_now = wait.until(EC.presence_of_element_located((By.XPATH,"//div[contains(@class,'cmbtv')]/button")))
    not_now.click()
    
Login("username","password")
```

# 4 Follow/Unfollow given handle

```python
def follow_unfollow(profile_name):
    
    # Opeing the Instagram Handle of “So Delhi” 
    
    Searching_opening(profile_name)
    
    #Following the given handle
    
    follow = wait.until(EC.presence_of_element_located((By.XPATH,"//button[contains(@class,'_5f5mN')]")))
    if follow.get_attribute("innerHTML") == "Following":
        print("Already Following")
    else:
        follow.click()
        print("Followed")
    time.sleep(2)
    
    #Unfollowing the given handle

    unfollow_trigger = driver.find_element_by_xpath("//button[contains(@class,'_5f5mN')]")
    unfollow_trigger.click()    # to unfollow 
    
    unfollow = driver.find_element_by_xpath("//button[contains(@class,'aOOlW')]") # gives the first occurence which is of unfollow button
    unfollow.click()
    print("Now unfollowed")
    driver.back()

    return

follow_unfollow("So Delhi") ## following/unfollowing of So Delhi
```

# 5 Like/Unlike posts

```python
def like_unlike(profile_name):
    Searching_opening(profile_name)
    #time.sleep(10)
    post = wait.until(EC.presence_of_element_located((By.CLASS_NAME,"_9AhH0")))
    post.click()
    
    #Liking the first 30 posts

    
    for i in range(30):
        
        #button = wait.until(EC.presence_of_element_located((By.XPATH,"//span[contains(@class,'glyphsSpriteHeart__')]")))
        p = wait.until(EC.presence_of_element_located((By.XPATH,'//span[@class = "fr66n"]/button/div/span')))
        ans = p.get_attribute('innerHTML').split('"')[1]
        if ans == "Like":
            p.click()
            print("Liked post ",i+1,)
        else:
            print("Post ",i+1," is already liked")
        
        next_button = driver.find_element_by_link_text("Next")
        next_button.click()
    close = driver.find_element_by_xpath('//div[contains(@class , "_2dDPU")]/div[3]/button')
    close.click()
    post = wait.until(EC.presence_of_element_located((By.CLASS_NAME,"_9AhH0")))
    post.click()
    
    #Unliking the first 30 posts
    
    for i in range(30):
        #button = wait.until(EC.presence_of_element_located((By.XPATH,"//span[contains(@class,'glyphsSpriteHeart__')]")))
        p = wait.until(EC.presence_of_element_located((By.XPATH,'//span[@class = "fr66n"]/button/div/span')))
        ans = p.get_attribute('innerHTML').split('"')[1]       
        if ans == "Unlike":
            p.click()
            print("Unliked the post ",i+1)
        else:
            print("Post ",i+1," is already unliked")
            
        next_button = driver.find_element_by_link_text("Next")
        next_button.click()            
    close = driver.find_element_by_xpath('//div[contains(@class , "_2dDPU")]/div[3]/button')
    close.click()
    back = wait.until(EC.presence_of_element_located((By.XPATH,"//img[contains(@class,'s4Iyt')]")))
    back.click()


like_unlike("dilsefoodie") # liking and unliking the first 30 posts of dilsefoodie
```

# 6 Extract list of followers

```python
# 6.1 Extract the usernames of the first 500 followers of ‘foodtalkindia’
import time
from selenium.webdriver.common.keys import Keys

def extract_followers(profile_name):
    followers_name=[]
    Searching_opening(profile_name)
    time.sleep(10)
    follower=driver.find_elements_by_class_name('-nal3 ')
    time.sleep(3)
    follower=follower[1]
    follower.click()
    time.sleep(5)
    print("Printing followers of ",end='')
    print(profile_name)
    print("----------------------------------")
    while True:
            scr = driver.find_element_by_xpath('//a[contains(@class,"FPmhX notranslate")]')
            followers_list = driver.find_elements_by_class_name('wFPL8')
            if len(followers_list) >= 50:
                break
            scr.send_keys(Keys.END)
            time.sleep(2) 

    for i in range(50):
        followers_name.append(followers_list[i].get_attribute('innerHTML'))
        print(str(i+1)+") ",followers_list[i].get_attribute('innerHTML'))
    cut=driver.find_element_by_xpath('//div[@class="WaOAr"]/button')
    cut.click()    
    return followers_name     
foodtalk_followers = extract_followers('foodtalkindia')
sodelhi_followers = extract_followers('sodelhi')
home=driver.find_element_by_xpath('//div[contains(@class,"_lz6s")]/div/div[1]/a/div/div')
home.click()
```

```python
# 7. Checking the story of coding.ninjas

wait = WebDriverWait(driver,17) # Here I have applied maximum limit as 17 sec as max duration for story is 17sec(For video) and 7 sec for photo
import time
def check_story(profile_name):
    
    Searching_opening(profile_name)
    cl = driver.find_element_by_xpath("//img[@class='_6q-tv']") 
    thing = driver.find_element_by_xpath("//canvas[@class = 'CfWVH']")
    #thing =  driver.find_element_by_xpath("//div[contains(@class_ , 'RR-M-')]/canvas")
    top = int(thing.get_attribute("style").split(";")[1][-3])
    
    width = thing.get_attribute("style").split(";")[3][8:11]
    if width == '150':
        print("No story")
    elif width == '166':
        print("Story present but already seen")
    elif width == '168':
        print("Story present and not seen yet")
        cl.click()
        print("Now seen the story")
        while True:
            try:
                next = wait.until(presence_of_element_located((By.XPATH,"//div[@class='coreSpriteRightChevron']")))
                next.click()
            except:
                break # ie. clicks on next until it is present, when not present, error will come and then it will come to except part and break
    driver.back()
    
check_story("coding.ninjas")
            
```
